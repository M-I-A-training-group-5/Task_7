```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function parameters
currentValue = 0.0         # Initial current value
targetValue = 10           # Target value
alpha = 0.04               # Alpha value for the smoothing function
duration = 10.0            # Total duration of the simulation (in seconds)
frequency = 50             # Sampling frequency (in Hz)
target_reached_time = None # Initialized with None
```

```python
time = np.linspace(0, duration, int(duration * frequency) + 1) # Create a time axis
current_values = []                                            # Lists to store values for plotting
tolerance = 1e-2                                               # Tolerance for equality check
```

```python
# Simulate the function
for t in time:
    currentValue = (1 - alpha) * currentValue + alpha * targetValue
    current_values.append(currentValue)
    if abs(currentValue - targetValue) < tolerance and target_reached_time is None:
        target_reached_time = t
```

```python
# Plot the current values
plt.figure(figsize=(10, 6))
plt.plot(time, current_values, label='Current Value')
plt.axhline(targetValue, color='red', linestyle='--', label='Target Value')
plt.xlabel('Time (seconds)')
plt.ylabel('Value')
plt.title('Value vs. Time')

# Mark when it reaches the target value
if target_reached_time is not None:
    plt.scatter(target_reached_time, targetValue, color='green', marker='o', label='Target Reached', zorder=5)

plt.legend()
plt.grid(True)
plt.show()
```
